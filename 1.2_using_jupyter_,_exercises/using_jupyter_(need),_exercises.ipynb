## 1 Quadratic Equations

### Introduction


![](https://upload.wikimedia.org/wikipedia/commons/thumb/f/f8/Polynomialdeg2.svg/500px-Polynomialdeg2.svg.png)

(_Image from the [Wikipedia page on Quadratic equations](https://en.wikipedia.org/wiki/Quadratic_equation)_)

The general form of a quadratic equation is:
$$
ax^2+bx+c=0
$$



### Solutions


Provided $ a \ne 0 $, we can use an elementary algebraic method called [completing the square](https://en.wikipedia.org/wiki/Completing_the_square) to show that a quadratic equation has the following solution:
$$
x = \dfrac{-b \pm \sqrt{b^2-4ac}}{2a}
$$


### Discriminant


The quantity $\Delta = b^2 - 4ac$ is called the **discriminant** of the equation and decides the nature of its solutions. The table below shows the various possibilities.

|**Discriminant**|**Roots**|
|---|---|
|$\Delta = b^2 - 4ac = 0$|A single solution of $-b/(2a)$|
|$\Delta = b^2 - 4ac \gt 0$|Two distinct solutions|
|$\Delta = b^2 - 4ac \lt 0$|No real solutions; both are complex.|


## **Exercise 2**
1. What is a Git **repository**?
    
    A git repository is a folder/storage space where Git tracks a project file and their entire history of changes. It contains the project files and information about past versions.

2. Git keeps track of **changes** in the files in our repository


3. What is a **commit**?
    
    A commit is a saved snapshot of the project at a specific point in time. It records the changes made to the file with a message describing what was changed.

4. What is a **push**?
    
    **Push** sends our commits to a remote repository such as GitHub

5. What is a **pull**?
    
    **Pull** downloads the latest changes from a remote repository and updates our local repository to match it

6. What is the difference beteen **GitHub** and **GitHub Desktop**?
    
    Github is an online platform that hosts Git repositories and allows collaboration, version control, and code sharing. 
    On the other hand, GitHub desktop is a desktop application that lets you use Git and interact with Github more easily without typing commands in the terminal.
