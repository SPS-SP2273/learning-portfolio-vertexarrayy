```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]

for index, deg in enumerate(temperatures_celsius):
    fahrenheit=temperatures_celsius[index] + 273
    celsius=temperatures_celsius[index]
    print(f"{celsius},{fahrenheit}")
```

## **Exercise 2**

```python
for letter in ['A', 'B', 'C']:
    for number in [1, 2, 3]:
        print(f'{letter}{number}', end='\t')
    print('\n')
```

```python
for number1 in range(1,6):
    print(f'{number1} :', end='\t')
    for number in range(1,6):
        print(number1*number, end='\t')     
    print('\n')
```
