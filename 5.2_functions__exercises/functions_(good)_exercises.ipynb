## **Exercise 1**

```python
Fahrenheit=Fahrenheit
Kelvin=Kelvin

def convert_celsius(temperature_celsius,target_scale=Fahrenheit):
    if target_scale==Fahrenheit:
        return (temperature_celsius*9/5)+32
    else:
        temp1=temperature_celsius+273.15
        return temp1
        

convert_celsius(8384,Kelvin)
```

## **Exercise 2**

```python
Celsius=Celsius

def convert_fahrenheit(temperature_fahrenheit,target_scale=Celsius):
    if target_scale==Celsius:
        temp2=f'{(temperature_fahrenheit-32)*5/9:.2f}'
        return temp2
    else:
        temp3=f'{(temperature_fahrenheit+459.67)*5/9:.2f}'
        return temp3

convert_fahrenheit(50,)
```

## **Exercise 3**

```python
def convert_temperature(temperature, source_scale, target_scale):
    if source_scale == target_scale:
        return temperature

    # Celsius as source
    if source_scale == "Celsius":
        if target_scale == "Fahrenheit":
            return convert_celsius(temperature, Fahrenheit)
        elif target_scale == "Kelvin":
            return convert_celsius(temperature, Kelvin)

    # Fahrenheit as source
    elif source_scale == "Fahrenheit":
        if target_scale == "Celsius":
            return convert_fahrenheit(temperature, Celsius)
        elif target_scale == "Kelvin":
            return convert_fahrenheit(temperature, Kelvin)

    # Kelvin as source
    elif source_scale == "Kelvin":
        if target_scale == "Celsius":
            return temperature - 273.15
        elif target_scale == "Fahrenheit":
            return f'{(temperature-273.15)*9/5+32:.2f}'
        
convert_temperature(5, "Kelvin", "Celsius")
```
