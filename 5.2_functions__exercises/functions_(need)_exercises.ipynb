## **Exercise 1**

```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
```

This function works without an else statement because a return immediately terminates the function. If the condition name == 'Batman' is true, the function returns the first greeting and exits. Otherwise, execution continues to the next line and returns the default greeting, which effectively serves as the else case.

```python
def integer(n):
    if n<0:
        return f"please type a positive integer"
    elif n==0:
        return f'1'
    else:
        int=1
        for i in range(1, n+1):
            int *= i
    return int

integer(5)
```

```python
def factorial(n):
    if n < 0:
        return "Invalid input"
    elif n == 0:
        return 1
    else:
        result = 1
        for number in range(1, n+1):
            result *= number
    return result

factorial(5)
```
