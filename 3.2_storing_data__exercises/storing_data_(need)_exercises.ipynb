## **Exercise 1**

1. Two similarities between lists and arrays.
    1. both list and array store multiple pieces of data in a single variable
    2. both list and array use  zzero-based indexing to access elements


2. Two differences between lists and arrays.
    1. Data type restriction:
        - Lists can contain elements of different data types (e.g. integers, floats, strings together).
        - arrays require all elements to be of the same data type.
    2. How operations work
        - Operations on lists usually act on the entire list as one object (e.g. adding or multiplying makes the list longer).
        - Operations on arrays act element-wise (e.g. adding or multiplying affects each element individually).

3. What is a dictionary?
    - A dictionary is a data structure that stores information as keyâ€“value pairs, where each key is used to access its associated values. This allows related pieces of information to be stored together in one structure, without relying on order.

## **Exercise 5**

```python
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

x1_array=np.array(x_1)
x2_array=np.array(x_2)

print(f"the mean of x_1 is {x1_array.mean()}")
print(f"the mean of x_2 is {x2_array.mean()}")
print(f"the standard deviation of x_1 is {x1_array.std()}")
print(f"the standard deviation of x_2 is {x2_array.std()}")

print(f"the r value using np.corrcoef is: {np.corrcoef(x1_array,x2_array)[0,1]}")

r = np.sum((x1_array - x1_array.mean()) * (x2_array - x2_array.mean())) / np.sqrt(np.sum((x1_array - x1_array.mean())**2) * np.sum((x2_array - x2_array.mean())**2))
print(r)
```

the values obtained from both np.corrcoef() is exactly the same as the one obtained by using np.sum() and np.sqrt()
the r value is very close to 1, indicating a strong positive correlation
