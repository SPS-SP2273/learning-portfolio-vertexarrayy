## **Exercise 3**

|  # |     Term     |                                            Description                                           |
|:--:|:------------:|:------------------------------------------------------------------------------------------------:|
|  1 |   Function   | A named piece of code that performs a specific task and may or may not take arguments.           |
|  2 |   Arguments  | Values passed into a function using parentheses to control how the function operates.            |
|  3 |   Comments   | Explanatory notes in the code that are ignored by Python and used to improve readability.        |
|  4 | Indentations | Spaces or tabs that Python uses to define blocks of code and control program structure.          |
|  5 |   Packages   | Collections of pre-written Python code that can be imported to extend Python’s functionality.    |
|  6 |       .      | Indicates ownership, allowing access to methods or attributes belonging to an object or package. |
|  7 |       :      | Marks the beginning of an indented code block such as in functions or conditional statements.    |
|  8 |      []      | Used to create lists and to access elements within lists through indexing.                       |
|  9 |   [ [],[] ]  | Represents a nested list, often used to store data in a table-like or matrix structure.          |
| 10 |      {}      | Used to create dictionaries that store data as key–value pairs.                                  |
